# Popularity Based Recommended System

```python
ratings_with_name=Ratings.merge(Books,on='ISBN')
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Rating'].reset_index()
num_rating_df
```

```python
avg_rating_df=ratings_with_name.groupby('Book-Title').mean()['Rating'].reset_index()
avg_rating_df.rename(columns={'Rating':'Avg_ratings'},inplace=True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['Rating']>=250].sort_values('Avg_ratings',ascending=False).head(50)
```

```python
Books.drop_duplicates('Book-Title').shape
```

```python
popular_df=popular_df.merge(Books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','Rating','Avg_ratings']]
```

# Collaboravtive Filtering Based Recommendation System

```python
ratings_with_name.groupby('User-ID').count()['Rating'] #we get ratings each user gave
```

```python
x=ratings_with_name.groupby('User-ID').count()['Rating']>200
x[x] #Here we get users who gave more than 200 ratings
```

```python
read_write_users=x[x].index
```

```python
filetered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(read_write_users)]
```

```python
filetered_rating #we r filtering based on users
```

```python
y=filetered_rating.groupby('Book-Title').count()['Rating']>=50 #grouping based on Book_Title
```

```python
famous_books=y[y].index #naming as famous books
```

```python
famous_books
```

```python
final_raings=filetered_rating[filetered_rating['Book-Title'].isin(famous_books)]
```

```python
final_raings.drop_duplicates()
```

```python
pt=final_raings.pivot_table(index='Book-Title',columns='User-ID',values='Rating')
```
